```python
# 리스트 컴프리핸션 (2차원 이상의 리스트를 초기화 할 때 사용)
# 원소를 8개 포함하는 1차원 리스트 초기화
arr = [5] * 8
print(arr)
```

```python
# 4 X 5 크기를 갖는 2차원 리스트 초기화
arr = [[0] * 5 for _ in range(4)]
print(arr)
```

```python
# 4 X 5 크기를 갖는 2차원 리스트 초기화
arr = [[i] * 5 for i in range(4)]
print(arr)
```

```python
# 4 X 5 크기를 갖는 2차원 리스트 초기화
arr = [[(i * 5) + j for j in range(5)] for i in range(4)]
print(arr)
```

```python
arr1 = ["컴퓨터", "키보드", "모니터"]
arr2 = ['computer', 'keyboard', 'monitor']

data = {}
print(arr1[1])
for i in range(3):
  data[arr1[i]] = arr2[i]
print(data)
```

```python
# 모든 키를 하나씩 확인할 때는 key 메서드 사용함
data = {}
data['apple'] = "사과"
data['banana'] = "바나나"
data['carrot'] = "당근"

for key in data.keys():
  print("key :", key, ", value:", data[key])
```

```python
a = {'name':'ezen', 'phone' : '01023439874', 'birth' : '990123'}
b = {0 : 'Hello Ezen!'}
c = {'arr' : [1, 2, 3, 4]}
```

```python
# key()
print('a :', a.keys())
print('b :', b.keys())
print('c :', c.keys())

print('a :', list(a.keys()))
print('b :', list(b.keys()))
print('c :', list(c.keys()))
```

```python
# values()
print('a :', a.values())
print('b :', b.values())
print('c :', c.values())

print('a :', list(a.values()))
print('b :', list(b.values()))
print('c :', list(c.values()))
```

```python
# items()
print('a :', a.items())
print('b :', b.items())
print('c :', c.items())

print('a :', list(a.items()))
print('b :', list(b.items()))
print('c :', list(c.items()))
```
