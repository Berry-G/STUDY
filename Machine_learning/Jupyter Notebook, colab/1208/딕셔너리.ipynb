```python
# 딕셔너리 자료형 (순서x, 중복x, 수정o, 삭제o)
"""
  1. 사전 자료형
    1) 데이터를 키와 값 쌍의 형태로 저장할 때 사용할 수 있음
    
    사전 데이터[키] = 값
"""
```

```python
# 선언
a = {'name':'ezen', 'phone' : '01023439874', 'birth' : '990123'}
b = {0 : 'Hello Ezen!'}
c = {'arr' : [1, 2, 3, 4]}

print('a :', type(a), a)
print('a :', type(b), b)
print('a :', type(c), c)
```

```python
# 출력
print('a :', a['name'])       # 존재 X => 에러발생  
print('a :', a.get('phone1')) # 존재 X => None 처리
print('b :', b[0])
print('b :', b.get(0))
print('c :', c['arr'])
print('c :', c['arr'][3])
print('c :', c.get('arr'))
```

```python
# 딕셔너리 추가
a['address'] = 'seoul'
print('a :', a)
```

```python
# rank [1, 2, 3]
a['rank'] = [1, 2, 3]
print('a :', a)
```
