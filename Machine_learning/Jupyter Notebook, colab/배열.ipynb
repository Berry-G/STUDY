```python
# 리스트 컴프리헨션 (2차원 이상의 리스트를 초기화 할 때 사용)
# 원소를 8개 포함하는 1차원 리스트 초기화
arr = [5] * 8
print(arr)
```

```python
# 4 X 5 크기를 갖는 2차원 리스트 초기화
arr = [[0] * 5 for _ in range(4)]
print(arr)
```
