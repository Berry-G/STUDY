- 사용자가 입력을 받는 두자리 이상 숫자 N에 대해서, 십의 자리 숫자를 출력하는 코드를 작성하시오.
- 예를 들어, 1234505055598838303394949494이 입력으로 들어오면 출력 결과는 9
- 8484758493727484838293847374이 입력으로 들어오면 출력 결과는 7
- 51이 입력으로 들어오면 출력 결과는 5

```python
twoDigits = input("두 자리 이상의 숫자 입력 : ")
print("십의 자리 숫자 :", twoDigits[-2])
```
