```python
"""
  머신러닝에서는 모델이 학습을 하게 되면, 학습에 주어진 인자들을 가지고 어떤 결과가 나왔는지 출력을 하게 됩니다.
  주어진 인자를 해당하는 자리에 위치하는 출력문을 작성하시오.
  아래는 출려문의 예시입니다.

  - Ezen777모델로 10 epoch를 반복 학습시킨 결과는 학습 정확도 : 0.85, 검증 정확도 : 0.83, 테스트 정확도 : 0.76으로 나왔습니다.
"""

# 다음 주어진 변수를 활용하여 코드를 작성하시오.
epoch = 10
train_accuracy = 0.85
val_accuracy = 0.83
test_accuracy = 0.76
model_name = "Ezen777"
```

```python
# case1
print(f"{model_name}모델로 {epoch} epoch를 반복 학습시킨 결과는 학습 정확도 : {train_accuracy}, 검증 정확도 : {val_accuracy}, 테스트 정확도 : {test_accuracy}으로 나왓습니다.")
```

```python
# case2
print("{}모델로 {} epoch를 반복 학습시킨 결과는 학습 정확도 : {}, 검증 정확도 : {}, 테스트 정확도 : {}으로 나왓습니다.".format(model_name, epoch, train_accuracy, val_accuracy, test_accuracy))
```

```python
# case3
print("%s모델로 %d epoch를 반복 학습시킨 결과는 학습 정확도 : %.2f, 검증 정확도 : %.2f, 테스트 정확도 : %.2f으로 나왓습니다."%(model_name, epoch, train_accuracy, val_accuracy, test_accuracy))
```

```python
"""
  어떤 반의 학생 5명의 키에 대한 정보가 주어집니다.
  이를 리스트로 만들고 heights라는 변수에 저장하시오.
  167, 160, 175.5, 184, 180
"""

heights = [167, 160, 175.5, 184, 180]
print(heights)
```

```python
# 학생 5명의 키가 작은 순부터 큰 순서대로 정렬하여 출력하시오.
heights.sort()
print(heights)
```

```python
# 위에서 만든 리스트이 데이터 타입을 확인해 보시오.
print(type(heights))
```

```python
# 위의 리스트의 데이터 길이를 출력하시오.
print(len(heights))
```

```python
# 학생들 중 키가 제일 컸던 학생이 전학을 가게 되었습니다. 이를 코드로 구혆시오.
# 위의 리스트의 데이터 길이를 출력하시오
print(max(heights))
heights.remove(max(heights))
print(heights)
print(len(heights))
```

```python
"""
  전학을 갔떤 친구보다 더 키가 큰 순신이라는 친구가 전학을 왔습니다.
  이 친구의 키는 200cm 입니다.
  이를 반영하여 코드를 구현하시오. heights의 길이를 구하시오.
"""
heights.append(200)
print(len(heights))
```

```python
"""
  학교에서 더이상 전학을 받지도, 보내지도 않기로 결정을 하였습니다.
  키 데이터 값이 더 이상 변경되지 않기를 원합니다.
  리소스의 값을 변경할수 없게 코드를 작성하시오.
"""

heights = tuple(heights)
print(heights)
print(type(heights))
```

```python
"""
  5명의 학생들의 키는 있지만, 누구의 키 인지 알 수 없습니다.
    최선혜(160), 전수정(167), 이규황(170), 반태희(175.5), 이경빈(200)
  데이터 타입 중 Dictionary로 위처럼 이름(name-key)과 키(height-value)를 같이 붙여주려고 합니다.
  이를 코드로 작성해서 class_A라는 변수에 저장하시오.
"""

class_A = {"최선혜" : 160, "전수정" : 167, "이규황" : (170), "반태희" : (175.5), "이경빈" : (200)}
print(class_A)
print(type(class_A))
print(class_A["반태희"])
```
