- 윤년 규칙 (그레고리 력)
  1. 4의 배수일것
  2. 100의 배수면 예외
  3. 400의 배수이면 윤년

```python
num = int(input("년도를 입력하세요 :"))
if((num % 4 == 0 and num%100 != 0) or num%400==0) :
  print(num, "년은 윤년입니다.")
else :
  print(num, "년은 윤년이 아닙닙니다.")
```

```python
sum1 = 0
cnt1 = 1

while cnt1 <= 100:
  sum1 += cnt1
  cnt1 += 1

print('1~100 합 :', sum1)
print('1~100 합 :', sum(range(1,101)))
print('1~100 안에 3의 배수의 합 :', sum(range(1, 101, 3)))
```
